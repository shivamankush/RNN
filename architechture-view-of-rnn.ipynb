```python
from keras import Sequential
```

```python
from keras.layers import Dense,SimpleRNN
```

```python
model=Sequential()
```

```python
# 3 nodes 
# Maximum 4 timestamp and input_features is 5
model.add(SimpleRNN(3,input_shape=(4,5)))
# Since we are working with Binary Classificatin we take Sigmoid function as Activation function
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
print(model.get_weights()[0].shape)
model.get_weights()[0]
```
